```python
import pandas as pd

df = pd.read_excel("para_data.xlsx")
df = df[['text', 'label']][:2166]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.20, stratify=df['label'], random_state=42
)
```

```python
vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# you change max features
```

```python

model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_vec, y_train)
lr_model=model

#iteration = 1000 for good results
```

```python
y_pred = model.predict(X_test_vec)

print("Classification Report:")
print(classification_report(y_test, y_pred, digits=4))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import roc_auc_score


y_pred_prob = model.predict_proba(X_test_vec)[:, 1]
auc_roc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC-ROC: {auc_roc}")
```

```python

cm = confusion_matrix(y_test, y_pred)

TP = cm[1, 1] 
FP = cm[0, 1] 
TN = cm[0, 0] 
FN = cm[1, 0] 

print(f"True Positives (TP): {TP}")
print(f"False Positives (FP): {FP}")
print(f"True Negatives (TN): {TN}")
print(f"False Negatives (FN): {FN}")
```

# Tweaking and Oversampling

```python
import json
import pandas as pd


with open('paragraphs.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

rows = [{'Paragraph Number': int(k), 'Paragraph': v['paragraph']} for k, v in data.items()]

df = pd.DataFrame(rows).sort_values('Paragraph Number')

df.to_excel('paragraphsbefore.xlsx', index=False, engine='openpyxl')
```

```python
import pandas as pd

df2 = pd.read_excel("paragraphsextra.xlsx")
df2 = df2[['Paragraph', 'Label']][:76]
```

```python
df2 = df2.rename(columns={
    'Paragraph': 'text',
    'Label': 'label'
})

df2
```

```python
import pandas as pd

merged_df = pd.concat([df, df2], ignore_index=True)

merged_df
```

```python
df=merged_df

X_train, X_test, y_train, y_test = train_test_split(
    df['text'], df['label'], test_size=0.2, stratify=df['label'], random_state=143
)

vectorizer = TfidfVectorizer(max_features=10000, ngram_range=(1, 2))
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train_vec, y_train)
lr_model=model


y_pred = model.predict(X_test_vec)

print("Classification Report:")
print(classification_report(y_test, y_pred, digits=4))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
